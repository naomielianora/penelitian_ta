```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style global
sns.set(style="whitegrid", palette="pastel", font_scale=1)

# Load CSV
df = pd.read_csv("hasil_perbandingan_metode1_6.csv")
df.columns = df.columns.str.strip()  # Bersihkan nama kolom
```

# Rata-rata Jarak Tiap Metode

```python
# Rata-rata Nilai Tiap Metode
average_methods = df[["M1", "M2", "M3", "M4", "M5", "M6"]].mean()
average_methods.index = ["Metode 1", "Metode 2", "Metode 3", "Metode 4", "Metode 5", "Metode 6"]

plt.figure(figsize=(10, 6))
ax2 = sns.barplot(x=average_methods.index, y=average_methods.values, 
                  order=["Metode 1", "Metode 2", "Metode 3", "Metode 4", "Metode 5", "Metode 6"],
                  palette=['skyblue', 'salmon', 'lightgreen', 'orange', 'violet', 'gray'])
plt.title("Rata-rata Jarak Tiap Metode", fontsize=16)
plt.xlabel("")
plt.ylabel("Rata-rata Jarak (km)")

# Menambahkan label pada tiap bar
for i, v in enumerate(average_methods.values):
    ax2.annotate(f'{v:.2f}', (i, v),
                 ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

# Distribusi Hasil Jarak Tiap Metode

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ubah data jadi long format
df_long = df[["M1", "M2", "M3", "M4", "M5", "M6"]].melt(
    var_name="Metode", value_name="Jarak"
)

# Ganti nama metode jadi "Metode 1" dst
df_long["Metode"] = df_long["Metode"].map({
    "M1": "Metode 1",
    "M2": "Metode 2",
    "M3": "Metode 3",
    "M4": "Metode 4",
    "M5": "Metode 5",
    "M6": "Metode 6"
})

# Plot boxplot
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x="Metode", y="Jarak", data=df_long, 
                 order=["Metode 1", "Metode 2", "Metode 3", "Metode 4", "Metode 5", "Metode 6"],
                 palette=['skyblue', 'salmon', 'lightgreen', 'orange', 'violet', 'gray'])

plt.title("Distribusi Jarak Tiap Metode", fontsize=16)
plt.xlabel("")
plt.ylabel("Jarak (km)")

plt.tight_layout()
plt.show()
```

# Distribusi Metode Terbaik

```python
# Distribusi Metode Terbaik
plt.figure(figsize=(10, 6))
ax1 = sns.countplot(
    data=df,
    x='Best Method',
    order=["Metode 1", "Metode 2", "Metode 3", "Metode 4", "Metode 5", "Metode 6"],
    palette=['skyblue', 'salmon', 'lightgreen', 'orange', 'violet', 'gray']
)

plt.title("Distribusi Frekuensi Pemilihan Metode Terbaik", fontsize=16)
plt.xlabel("")
plt.ylabel("Jumlah Gunung")

# Menambahkan label pada tiap bar
for p in ax1.patches:
    height = p.get_height()
    ax1.annotate(f'{int(height)}',
                 (p.get_x() + p.get_width() / 2, height),
                 ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Distribusi Jarak Berdasarkan Best Method
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df,
    x='Best Method',
    y='Min',
    palette='pastel',
    order=["Metode 1", "Metode 2", "Metode 3", "Metode 4", "Metode 5", "Metode 6"]
)
plt.title("Distribusi Jarak dari Metode yang Terpilih sebagai Terbaik", fontsize=16)
plt.xlabel("Metode Terbaik")
plt.ylabel("Jarak (km)")
plt.tight_layout()
plt.show()
```
