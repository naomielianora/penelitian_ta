```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style global
sns.set(style="whitegrid", palette="pastel", font_scale=1)
```

# Rata-rata Jarak Kesembilan Metode

```python
# Load CSV
df = pd.read_csv("hasil_perbandingan_9 metode.csv")
df.columns = df.columns.str.strip()  # Bersihkan nama kolom

# Rata-rata Nilai Tiap Metode (9 metode)
average_methods = df[["M1", "M2", "M3", "M4", "M5", "M6", "M7", "M8", "M9"]].mean()
average_methods.index = [
    "Metode 1", "Metode 2", "Metode 3", 
    "Metode 4", "Metode 5", "Metode 6", 
    "Metode 7", "Metode 8", "Metode 9"
]

plt.figure(figsize=(12, 6))
ax2 = sns.barplot(
    x=average_methods.index, 
    y=average_methods.values, 
    order=[
        "Metode 1", "Metode 2", "Metode 3", 
        "Metode 4", "Metode 5", "Metode 6", 
        "Metode 7", "Metode 8", "Metode 9"
    ],
    palette=['skyblue', 'salmon', 'lightgreen', 'orange', 'violet', 'gray', 'gold', 'turquoise', 'pink']
)
plt.title("Rata-rata Jarak Tiap Metode", fontsize=16)
plt.xlabel("")
plt.ylabel("Rata-rata Jarak (km)")

# Menambahkan label pada tiap bar
for i, v in enumerate(average_methods.values):
    ax2.annotate(f'{v:.2f}', (i, v),
                 ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

```

# Analisis Metode Pilihan

```python
# Load CSV
df = pd.read_csv("hasil_perbandingan_selected methods.csv")
df.columns = df.columns.str.strip()  # Bersihkan nama kolom
```

## Rata-rata Jarak

```python
# --- Rata-rata nilai tiap metode (Metode 1, 2, 4, 8, 9) ---
average_methods = df[["Metode 1", "Metode 2", "Metode 4", "Metode 8", "Metode 9"]].mean()
average_methods.index = ["Metode 1", "Metode 2", "Metode 4", "Metode 8", "Metode 9"]

# --- Plot bar rata-rata ---
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x=average_methods.index,
    y=average_methods.values,
    order=["Metode 1", "Metode 2", "Metode 4", "Metode 8", "Metode 9"],
    palette=['skyblue', 'salmon', 'lightgreen', 'orange', 'violet']
)

plt.title("Rata-rata Jarak Tiap Metode (Metode Terpilih)", fontsize=16)
plt.xlabel("")
plt.ylabel("Rata-rata Jarak (km)")

# --- Tambahkan label angka di atas bar ---
for i, v in enumerate(average_methods.values):
    ax.annotate(f'{v:.2f}', (i, v),
                ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

## Distribusi Hasil Jarak Tiap Metode

```python
# Ubah data jadi long format
df_long = df[["Metode 1", "Metode 2", "Metode 4", "Metode 8", "Metode 9"]].melt(
    var_name="Metode", value_name="Jarak"
)

# Plot boxplot
plt.figure(figsize=(10, 6))
ax = sns.boxplot(
    x="Metode", 
    y="Jarak", 
    data=df_long,
    order=["Metode 1", "Metode 2", "Metode 4", "Metode 8", "Metode 9"],
    palette=['skyblue', 'salmon', 'lightgreen', 'orange', 'violet']
)

plt.title("Distribusi Jarak Tiap Metode (Metode Terpilih)", fontsize=16)
plt.xlabel("")
plt.ylabel("Jarak (km)")

plt.tight_layout()
plt.show()
```

## Distribusi Metode Terbaik

```python
# Distribusi Metode Terbaik
plt.figure(figsize=(10, 6))
ax1 = sns.countplot(
    data=df,
    x='Best Method',
    order=["Metode 1", "Metode 2", "Metode 4", "Metode 8", "Metode 9"],
    palette=['skyblue', 'salmon', 'lightgreen', 'orange', 'violet']
)

plt.title("Distribusi Frekuensi Pemilihan Metode Terbaik (Metode Terpilih)", fontsize=16)
plt.xlabel("")
plt.ylabel("Jumlah Gunung")

# Menambahkan label pada tiap bar
for p in ax1.patches:
    height = p.get_height()
    ax1.annotate(f'{int(height)}',
                 (p.get_x() + p.get_width() / 2, height),
                 ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Distribusi Jarak Berdasarkan Best Method
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df,
    x='Best Method',
    y='Min',
    palette='pastel',
    order=["Metode 1", "Metode 2", "Metode 4", "Metode 8", "Metode 9"]
)
plt.title("Distribusi Jarak dari Metode yang Terpilih sebagai Terbaik", fontsize=16)
plt.xlabel("Metode Terbaik")
plt.ylabel("Jarak (km)")
plt.tight_layout()
plt.show()

```

## Distribusi Metode Terburuk

```python
plt.figure(figsize=(10, 6))
ax1 = sns.countplot(
    data=df,
    x='Worst Method',
    order=["Metode 1", "Metode 2", "Metode 4", "Metode 8", "Metode 9"],
    palette=['skyblue', 'salmon', 'lightgreen', 'orange', 'violet']
)

plt.title("Distribusi Metode Terburuk (Metode Terpilih)", fontsize=16)
plt.xlabel("")
plt.ylabel("Jumlah Gunung")

# Menambahkan label pada tiap bar
for p in ax1.patches:
    height = p.get_height()
    ax1.annotate(f'{int(height)}',
                 (p.get_x() + p.get_width() / 2, height),
                 ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df,
    x='Worst Method',
    y='Max',
    palette='pastel',
    order=["Metode 1", "Metode 2", "Metode 4", "Metode 8", "Metode 9"]
)
plt.title("Distribusi Jarak dari Metode yang Terpilih sebagai Terburuk", fontsize=16)
plt.xlabel("Metode Terburuk")
plt.ylabel("Jarak (km)")
plt.tight_layout()
plt.show()
```

## Kesimpulan

- **Rata-rata Jarak**
  - Metode 8 memiliki rata-rata jarak terkecil (0.56 km), disusul oleh Metode 9 (0.67 km)  
  - Metode 2 menghasilkan rata-rata jarak terbesar (1.14 km)  

- **Sebaran Jarak**
  - Metode 8 menunjukkan distribusi jarak yang paling rendah dan stabil  
  - Metode 2 cenderung memiliki sebaran yang lebih lebar dengan banyak outlier, menandakan hasil yang kurang konsisten  

- **Frekuensi Terpilih sebagai Metode Terbaik**
  - Metode 8 paling sering menjadi yang terbaik (25 gunung)  
  - Diikuti Metode 9 (13 gunung) dan Metode 4 (12 gunung)  
  - Metode 1 dan 2 relatif jarang terpilih  

- **Sebaran Jarak dari Metode yang Terpilih sebagai Terbaik**
  - Saat suatu metode dipilih sebagai terbaik, jarak yang dihasilkan umumnya < 1 km  
  - Metode 8 kembali menonjol karena jaraknya paling kecil dan stabil  

- **Distribusi Metode Terburuk**
  - Metode 2 paling sering menjadi yang terburuk (25 gunung)  
  - Metode 8 hampir tidak pernah menjadi yang terburuk (hanya 3 gunung)  
  - Ini menegaskan keunggulan konsistensi Metode 8 dibanding yang lain  

- **Sebaran Jarak dari Metode yang Terpilih sebagai Terburuk**
  - Metode 2 menunjukkan jarak terburuk yang jauh lebih besar, bahkan hingga > 6 km  
  - Sebaliknya, Metode 9 dan Metode 1 tetap relatif rendah meskipun menjadi yang terburuk  

## Inti Kesimpulan
Metode 8 secara konsisten memberikan hasil paling baik, dengan rata-rata jarak terkecil, distribusi paling stabil, dan frekuensi paling sering terpilih sebagai metode terbaik. Sebaliknya, Metode 2 cenderung menjadi metode yang paling tidak andal karena sering menjadi yang terburuk dan memiliki rata-rata jarak terbesar.
