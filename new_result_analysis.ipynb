```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style global
sns.set(style="whitegrid", palette="pastel", font_scale=1)

# Load CSV
df = pd.read_csv("hasil_perbandingan_metode4_6.csv")
df.columns = df.columns.str.strip()  # Bersihkan nama kolom
```

# Rata-rata Jarak Tiap Metode

```python
# Rata-rata Nilai Tiap Metode (Metode 4–6)
average_methods = df[["M4", "M5", "M6"]].mean()
average_methods.index = ["Metode 4", "Metode 5", "Metode 6"]

plt.figure(figsize=(8, 5))
ax2 = sns.barplot(
    x=average_methods.index,
    y=average_methods.values,
    order=["Metode 4", "Metode 5", "Metode 6"],
    palette=['orange', 'violet', 'gray']
)
plt.title("Rata-rata Jarak Tiap Metode (4–6)", fontsize=16)
plt.xlabel("")
plt.ylabel("Rata-rata Jarak (km)")

# Menambahkan label pada tiap bar
for i, v in enumerate(average_methods.values):
    ax2.annotate(f'{v:.2f}', (i, v),
                 ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

# Distribusi Hasil Jarak Tiap Metode

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ubah data jadi long format hanya untuk metode 4–6
df_long = df[["M4", "M5", "M6"]].melt(
    var_name="Metode", value_name="Jarak"
)

# Ganti nama metode jadi "Metode 4" dst
df_long["Metode"] = df_long["Metode"].map({
    "M4": "Metode 4",
    "M5": "Metode 5",
    "M6": "Metode 6"
})

# Plot boxplot
plt.figure(figsize=(8, 5))
ax = sns.boxplot(
    x="Metode", y="Jarak", data=df_long,
    order=["Metode 4", "Metode 5", "Metode 6"],
    palette=['orange', 'violet', 'gray']
)

plt.title("Distribusi Jarak Tiap Metode (4–6)", fontsize=16)
plt.xlabel("")
plt.ylabel("Jarak (km)")

plt.tight_layout()
plt.show()
```

# Distribusi Metode Terbaik

```python
# Distribusi Metode Terbaik (Metode 4–6)
plt.figure(figsize=(8, 5))
ax1 = sns.countplot(
    data=df,
    x='Best Method',
    order=["Metode 4", "Metode 5", "Metode 6"],
    palette=['orange', 'violet', 'gray']
)

plt.title("Distribusi Frekuensi Pemilihan Metode Terbaik (4–6)", fontsize=16)
plt.xlabel("")
plt.ylabel("Jumlah Gunung")

# Menambahkan label pada tiap bar
for p in ax1.patches:
    height = p.get_height()
    ax1.annotate(f'{int(height)}',
                 (p.get_x() + p.get_width() / 2, height),
                 ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Filter hanya Metode 4 - 6
df_filtered = df[df['Best Method'].isin(["Metode 4", "Metode 5", "Metode 6"])]

# Plot boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(
    data=df_filtered,
    x='Best Method',
    y='Min',
    palette='pastel',
    order=["Metode 4", "Metode 5", "Metode 6"]
)
plt.title("Distribusi Jarak dari Metode 4–6", fontsize=16)
plt.xlabel("Metode Terbaik")
plt.ylabel("Jarak (km)")
plt.tight_layout()
plt.show()
```

# Distribusi Metode Terburuk

```python
# Filter hanya Metode 4 - 6
df_filtered = df[df['Worst Method'].isin(["Metode 4", "Metode 5", "Metode 6"])]

# Plot countplot
plt.figure(figsize=(8, 6))
ax1 = sns.countplot(
    data=df_filtered,
    x='Worst Method',
    order=["Metode 4", "Metode 5", "Metode 6"],
    palette=['orange', 'violet', 'gray']
)

plt.title("Distribusi Metode Terburuk (Hanya 4–6)", fontsize=16)
plt.xlabel("Metode Terburuk")
plt.ylabel("Jumlah Gunung")

# Tambahkan label di atas bar
for p in ax1.patches:
    height = p.get_height()
    ax1.annotate(f'{int(height)}',
                 (p.get_x() + p.get_width() / 2, height),
                 ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Filter hanya Metode 4 - 6
df_filtered = df[df['Worst Method'].isin(["Metode 4", "Metode 5", "Metode 6"])]

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df_filtered,
    x='Worst Method',
    y='Max',
    palette=['orange', 'violet', 'gray'],
    order=["Metode 4", "Metode 5", "Metode 6"]
)

plt.title("Distribusi Jarak dari Metode yang Terpilih sebagai Terburuk (4–6)", fontsize=16)
plt.xlabel("Metode Terburuk")
plt.ylabel("Jarak (km)")
plt.tight_layout()
plt.show()
```
